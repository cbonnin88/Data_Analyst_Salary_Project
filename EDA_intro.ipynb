```python
import ast
import pandas as pd
from datasets import load_dataset
import ast
from matplotlib import pyplot as plt
import seaborn as sns


dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_da_fr = df[(df["job_country"] == "France") & (df["job_title_short"] == "Data Analyst")]
```

```python
df_da_fr
```

```python
df_plot = df_da_fr["job_location"].value_counts().head(10).to_frame()
```

```python
# The count of locations in France for Data Analysts :

sns.set_theme(style="ticks")
sns.barplot(data=df_plot, x="count", y="job_location", hue="count", palette="dark:g_r",legend=False)
sns.despine()
plt.title("Counts of Job Locations for Data Analysts in France")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```

```python
# Percentage of work from home positions and Degree Requirements
dict_column = {
    "job_work_from_home": "Work From Home",
    "job_no_degree_mention": "Job Degree Req.",
}

fig, ax = plt.subplots(1,2)
fig.set_size_inches(12, 5)

for key, (column,title) in enumerate(dict_column.items()):
    ax[key].pie(df_da_fr[column].value_counts(), startangle=90,autopct='%1.1f%%', labels=["False", "True"])
    ax[key].set_title(title)
    
plt.show()
```

```python
df_plot = df_da_fr["company_name"].value_counts().head(10).to_frame()
sns.set_theme(style="ticks")
sns.barplot(data=df_plot, x="count", y="company_name", hue="count", palette="dark:r_r", legend=False)
sns.despine()
plt.title("Counts of Companies for Data Analysts in France")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```
